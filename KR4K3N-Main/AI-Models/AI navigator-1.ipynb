```python
columns_to_select = ['World.PrimaryBattleship.x', 'World.PrimaryBattleship.y',
       'World.PrimaryBattleship.actions', 'World.PrimaryBattleship.heading',
       'World.PrimaryBattleship.desired_heading',
       'World.PrimaryBattleship.option_port',
       'World.PrimaryBattleship.option_starboard',
       'World.PrimaryBattleship.chosen_direction','World.PrimaryBattleship.out_of_bounds', 'World.PrimaryBattleship.Navigation.waypoints','World.PrimaryBattleship.RadarSonar.collision_warning',
       'World.PrimaryBattleship.RadarSonar.collision_event']
df_new = df1[columns_to_select].copy()
```

```python
df_new.head()
```

Feature extraction 

```python
dest_coordinates = df_new['World.PrimaryBattleship.Navigation.waypoints'][0]
dest_coordinates_list = ast.literal_eval(dest_coordinates)

```

```python
def calc_distance_from_dest(x,y,point):
    dx = x - point[0]
    dy = y - point[1]
    return math.sqrt(dx**2 + dy**2)

for i, point in enumerate(dest_coordinates_list):
    df_new[f'distance_to_point_{i + 1}'] = df_new.apply(lambda row: calc_distance_from_dest(row['World.PrimaryBattleship.x'], row['World.PrimaryBattleship.y'], point), axis=1)
```

```python
df_new.head()
```

```python
df_new = pd.get_dummies(df_new, columns=['World.PrimaryBattleship.actions'], prefix=['action'])
```

```python
df_new = pd.get_dummies(df_new, columns=['World.PrimaryBattleship.chosen_direction'], prefix=['direction'])
```

```python
df_new['World.PrimaryBattleship.RadarSonar.collision_warning'] = df_new['World.PrimaryBattleship.RadarSonar.collision_warning'].astype(int)
```

```python
df_new['World.PrimaryBattleship.RadarSonar.collision_event'] = df_new['World.PrimaryBattleship.RadarSonar.collision_event'].astype(int)
```

```python
df_new['World.PrimaryBattleship.out_of_bounds'] = df_new['World.PrimaryBattleship.out_of_bounds'].astype(int)
```

```python
df_new.head()
```

```python
y = df_new['World.PrimaryBattleship.heading']
```

```python
df_new.columns
```

```python
columns_to_drop = ['World.PrimaryBattleship.heading','World.PrimaryBattleship.Navigation.waypoints' ]
df_new = df_new.drop(columns=columns_to_drop)
```

```python
df_new.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Assuming you have a target variable 'y' for regression

# Split the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(df_new, y, test_size=0.2, random_state=42)

# Create a decision tree regressor
regressor = DecisionTreeRegressor()

# Fit the model on the training data
regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred = regressor.predict(X_test)

# Calculate the mean squared error (MSE) to evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
comparison_df = pd.DataFrame({'Actual (y_test)': y_test, 'Predicted (y_pred)': y_pred})
```
